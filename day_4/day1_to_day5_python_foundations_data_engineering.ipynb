### 5. Dynamic typing

Python assigns types at runtime based on the value.


```python
x = 100
print(x, type(x))

x = 100.5
print(x, type(x))

x = "completed"
print(x, type(x))

x = False
print(x, type(x))
```

### 25. Processing multiple records with a for loop


```python
orders = [
    {"order_id": "1001", "amount": "250.50", "status": "completed"},
    {"order_id": "1002", "amount": "-20", "status": "pending"},
    {"order_id": "1003", "amount": "499.99", "status": "completed"}
]

for order in orders:
    order_id = int(order["order_id"])
    amount = float(order["amount"])
    status = order["status"]

    if amount > 0 and status == "completed":
        print("Clean order:", {"order_id": order_id, "amount": amount})
```

## Data Engineering Examples Across Day 1 to Day 5


### 31. Example: Mock third-party order API


```python
import random

def get_order_from_api():
    orders = [
        {"order_id": "7001", "user_id": "501", "amount": "299.50", "status": "completed"},
        {"order_id": "7002", "user_id": "502", "amount": "-50", "status": "completed"},
        {"order_id": "7003", "user_id": "503", "amount": "450", "status": "pending"},
        {"order_id": "7004", "user_id": "504", "amount": "199.99", "status": "completed"}
    ]
    return random.choice(orders)

print(get_order_from_api())
```

### 32. Example: Preprocess one order from API


```python
order = get_order_from_api()

order_id = int(order["order_id"])
user_id = int(order["user_id"])
amount = float(order["amount"])
status = order["status"]

is_valid = amount > 0 and status == "completed"

if is_valid:
    tax = amount * 0.18
    final_amount = amount + tax
    print({
        "order_id": order_id,
        "user_id": user_id,
        "amount": amount,
        "tax": round(tax, 2),
        "final_amount": round(final_amount, 2)
    })
else:
    print("Invalid order skipped")
```

### 33. Example: Process multiple API records


```python
valid_orders = []
invalid_count = 0

for _ in range(10):
    order = get_order_from_api()

    order_id = int(order["order_id"])
    user_id = int(order["user_id"])
    amount = float(order["amount"])
    status = order["status"]

    if amount > 0 and status == "completed":
        valid_orders.append({
            "order_id": order_id,
            "user_id": user_id,
            "amount": amount
        })
    else:
        invalid_count += 1

print("Valid Orders:")
print(valid_orders)
print("Invalid Count:", invalid_count)
```

### 34. Example: Total revenue calculation


```python
total_revenue = 0

for order in valid_orders:
    total_revenue += order["amount"]

print("Total Revenue:", total_revenue)
```

### 35. Example: Mock trade API


```python
def generate_trade():
    symbols = ["AAPL", "TSLA", "GOOG"]
    sides = ["BUY", "SELL"]

    trade = {
        "trade_id": random.randint(1000, 9999),
        "symbol": random.choice(symbols),
        "side": random.choice(sides),
        "quantity": random.randint(1, 100),
        "price": round(random.uniform(100, 500), 2)
    }

    return trade

print(generate_trade())
```

### 36. Example: Trade preprocessing


```python
trade = generate_trade()

trade_id = trade["trade_id"]
symbol = trade["symbol"]
side = trade["side"]
quantity = trade["quantity"]
price = trade["price"]

trade_value = quantity * price

if quantity > 0 and price > 0 and (side == "BUY" or side == "SELL"):
    print({
        "trade_id": trade_id,
        "symbol": symbol,
        "side": side,
        "quantity": quantity,
        "price": price,
        "trade_value": round(trade_value, 2)
    })
else:
    print("Invalid trade")
```

### 37. Example: Process multiple trades


```python
total_trade_value = 0
buy_count = 0
sell_count = 0

for _ in range(5):
    trade = generate_trade()

    quantity = trade["quantity"]
    price = trade["price"]
    side = trade["side"]

    if quantity > 0 and price > 0:
        total_trade_value += quantity * price

        if side == "BUY":
            buy_count += 1
        elif side == "SELL":
            sell_count += 1

print("Total Trade Value:", round(total_trade_value, 2))
print("BUY Count:", buy_count)
print("SELL Count:", sell_count)
```

### 38. Example: Mock weather reading API


```python
def get_weather_reading():
    readings = [
        {"city": "Delhi", "temperature": "34.5", "humidity": "65", "status": "ok"},
        {"city": "Mumbai", "temperature": "29.0", "humidity": "75", "status": "ok"},
        {"city": "Chennai", "temperature": "31.2", "humidity": "80", "status": "ok"},
        {"city": "Kolkata", "temperature": "-200", "humidity": "90", "status": "error"}
    ]
    return random.choice(readings)

print(get_weather_reading())
```

### 39. Example: Weather data validation


```python
reading = get_weather_reading()

city = reading["city"]
temperature = float(reading["temperature"])
humidity = int(reading["humidity"])
status = reading["status"]

if temperature >= -50 and temperature <= 60 and humidity >= 0 and humidity <= 100 and status == "ok":
    category = "Hot" if temperature >= 30 else "Pleasant"
    print({
        "city": city,
        "temperature": temperature,
        "humidity": humidity,
        "category": category
    })
else:
    print("Invalid weather record")
```
