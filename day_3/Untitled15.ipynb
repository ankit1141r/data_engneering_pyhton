```python
import random
import time
def generate_trade():
    symbols = ["AAPL", "TSLA", "GOOG"]
    sides = ["BUY", "SELL"]

    trade = {
        "trade_id": random.randint(1000, 9999),
        "symbol": random.choice(symbols),
        "side": random.choice(sides),
        "quantity": str(random.randint(1, 100)),
        "price": str(round(random.uniform(100, 500), 2))
    }

    return trade
```

```python
def process_trades_for_one_minute():
    buy = 0
    sell = 0
    start_time = time.time()

    while time.time() - start_time < 60:  # run for 1 minute
        trade = generate_trade()  # single trade dict
        print(trade)
        time.sleep(2)

        # Calculate trade value = quantity × price
        quantity = int(trade["quantity"])
        price = float(trade["price"])
        trade_value = quantity * price

        if trade["side"] == "BUY":
            buy += trade_value
        elif trade["side"] == "SELL":
            sell += trade_value

    # After 1 minute, calculate result
    if buy >= sell:
        print("loss", buy - sell)
    else:
        print("profit", sell - buy)

# Run it
process_trades_for_one_minute()
```
